```python
combine_df = combine_df_2020.append(combine_df_2019,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combine_df_2018,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combine_df_2017,ignore_index=True,verify_integrity=True)
combine_df = combine_df.append(combine_df_2016,ignore_index=True,verify_integrity=True)
combine_df = combine_df.rename(columns = {'Player': 'player'})
combine_df = combine_df.drop(columns = {'Pos'})
combine_df
```

```python
draft_df = draft_df_2020.append(draft_df_2019,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draft_df_2018,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draft_df_2017,ignore_index=True,verify_integrity=True)
draft_df = draft_df.append(draft_df_2016,ignore_index=True,verify_integrity=True)
draft_df = draft_df.rename(columns = {"Pos.": "Pos", "College": "School", "Conf.": "Conf", "Rnd.": "Rnd","Player":"player"})
draft_df = draft_df.drop(columns = ['Pos'])
draft_df
```

```python
player = combine_df['player']
school = combine_df['School']

combine_player_df = {"player":player,
                "school":school}
combine_player_df = pd.DataFrame(combine_player_df)
combine_player_df
```

```python
player = draft_df['player']
school = draft_df['School']

draft_player_df = {"player":player,
                "school":school}
draft_player_df = pd.DataFrame(draft_player_df)
draft_player_df
```

```python
combined_df = pd.concat([draft_player_df, combine_player_df])
combined_df.tail()
```

```python
players_df = combined_df.drop_duplicates(subset= None, keep='first')
players_df
```

```python
# Connect to local database
# Pior: Create a python file called info in your local repo with user, password and database name.

# from secrets import user, password, database
rds_connection_string = f'{username}:{password}@localhost:5432/nfl_db'
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check the tables names to make sure where are we posting
engine.table_names()
```

```python
# Append data using pandas. name stands for table name, change it if neccesary.
players_df.to_sql(name='player', con=engine, if_exists='append', index=False)
```
