## Scrap the data from the web using pandas

```python
#url to 5 years period combine stats

url2020 = 'https://www.pro-football-reference.com/draft/2020-combine.htm#combine'
url2019 = 'https://www.pro-football-reference.com/draft/2019-combine.htm'
url2018 = 'https://www.pro-football-reference.com/draft/2018-combine.htm'
url2017 = 'https://www.pro-football-reference.com/draft/2017-combine.htm'
url2016 =  'https://www.pro-football-reference.com/draft/2016-combine.htm'
url2015 = 'https://www.pro-football-reference.com/draft/2015-combine.htm'
```

```python
#tables for 5 years period
tables2020 = pd.read_html(url2020)
tables2019 = pd.read_html(url2019)
tables2018 = pd.read_html(url2018)
tables2017 = pd.read_html(url2017)
tables2016 = pd.read_html(url2016)
```

# Clean the data using pandas
* Convert the data into a DataFrame
* Delete (drop) the columns we are not using: 'College' and 'Drafted (tm/rnd/yr)'

```python
# Convert the tables to a DataFrame
df_2020 = tables2020[0]
df_2019 = tables2019[0]
df_2018 = tables2018[0]
df_2017 = tables2017[0]
df_2016 = tables2016[0]
```

```python
# Clean 2020 DataFrame
df2020 = df_2020.drop(columns = 'Drafted (tm/rnd/yr)')
df2020 = df2020.drop(columns = 'College')
df2020
```

```python
# Clean 2019 DataFrame
df2019 = df_2019.drop(columns = 'Drafted (tm/rnd/yr)')
df2019 = df2019.drop(columns = 'College')
df2019
```

```python
# Clean 2018 DataFrame
df2018 = df_2018.drop(columns = 'Drafted (tm/rnd/yr)')
df2018 = df2018.drop(columns = 'College')
df2018
```

```python
# Clean 2017 DataFrame
df2017 = df_2017.drop(columns = 'Drafted (tm/rnd/yr)')
df2017 = df2017.drop(columns = 'College')
df2017
```

```python
# Clean 2016 DataFrame
df2016 = df_2016.drop(columns = 'Drafted (tm/rnd/yr)')
df2016 = df2016.drop(columns = 'College')
df2016
```

## Append the data into a single DataFrame to post it to the database

```python
# Add a year column for data storage porpuses
df2020['year']='2020'
df2019['year']='2019'
df2018['year']='2018'
df2017['year']='2017'
df2016['year']='2016'
df2020
```

```python
# Append the 5 years DataFrames into one for storage porpuses
df = df2020.append(df2019,ignore_index=True,verify_integrity=True)
df = df.append(df2018,ignore_index=True,verify_integrity=True)
df = df.append(df2017,ignore_index=True,verify_integrity=True)
df = df.append(df2016,ignore_index=True,verify_integrity=True)

df
```
